```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from  sklearn.model_selection import train_test_split 
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# Checking for overlap between train and test sets
train_indices = xtrain.index
test_indices = xtest.index
overlap = train_indices.intersection(test_indices)
print(f"Overlap: {len(overlap)}")
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
lr1=LogisticRegression(class_weight='balanced', random_state=1)
lr1.fit(xtrain,ytrain)
ypred=lr1.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(ytest,ypred)
mse=mean_squared_error(ytest,ypred)
rmse=np.sqrt(mse)
Accuracy=r2_score(ytest,ypred)
print(f"mae:{mae}\nmse:{mse}\nrsme:{rmse}\nAccuracy:{Accuracy}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```
